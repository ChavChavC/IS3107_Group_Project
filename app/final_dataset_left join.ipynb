```python
import pandas as pd

data1 = pd.read_csv('./data/agoda_booking_combined.csv')

data2 = pd.read_csv('./google_place_api_summary.csv')
```

```python
data2 = data2.rename(columns={'Hotel Name': 'hotel_name'})
```

```python
merged = pd.merge(data1, data2, on='hotel_name', how='left')
merged
```

```python
merged['Geometry']
```

```python
import numpy as np

def extract_lat_lng(row):
    if pd.isna(row):
        return pd.Series([np.nan, np.nan])
    else:
        lat_lng = eval(row)  # Convert string to dictionary
        return pd.Series([lat_lng['lat'], lat_lng['lng']])

# Apply the function to each row
merged[['Latitude', 'Longitude']] = merged['Geometry'].apply(extract_lat_lng)

# Drop the original 'Location' column if necessary
merged = merged.drop(columns=['Geometry'])
```

```python
merged.to_csv('./data/merged_final_draft.csv', index=False)
```
