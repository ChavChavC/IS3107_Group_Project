```python
import pandas as pd
import json
from pandas import json_normalize
import os
```

```python
# Read the file line by line and parse each line as JSON
data = []
with open(r'../raw/review.txt', 'r') as file:
    for line in file:
        try:
            json_obj = json.loads(line.strip())
            data.append(json_obj)
        except json.JSONDecodeError:
            print(f"Failed to parse JSON: {line}")
```

```python
# Flatten the JSON data
df = pd.json_normalize(data)
```

```python
# Read the file line by line and parse each line as JSON
data = []
with open(r'../raw/offering.txt', 'r') as file:
    for line in file:
        try:
            json_obj = json.loads(line.strip())
            data.append(json_obj)
        except json.JSONDecodeError:
            print(f"Failed to parse JSON: {line}")
```

```python
# Flatten the JSON data
offering_df = pd.json_normalize(data)
```

```python
offering_df
```

```python
offering_df.columns
```

```python
merged_df = pd.merge(df, offering_df[["id", "name", "address.region"]], left_on='offering_id', right_on="id", how='inner')
```

```python
print(len(df))
print(len(merged_df))
```

```python
merged_df.head(1)
```

```python
summary_df = merged_df.groupby('name').agg(
    overall_rating=('ratings.overall', 'mean'),
    total_num_of_reviews=('offering_id', 'count'),
    region=('address.region', lambda x: x.mode().iloc[0] if not x.empty else None),
    # region_shortform=('address.locality', lambda x: x.mode().iloc[0] if not x.empty else None),
    cleanliness=('ratings.cleanliness', 'mean'),
    service=('ratings.service', 'mean'),
    location=('ratings.location', 'mean'),  # If this column does not exist, it will be filled with NaN
    # facilities=('details', lambda x: x.mode().iloc[0] if not x.empty else None),  # Assuming 'details' represents facilities
    value=('ratings.value', 'mean')
).reset_index()
```

```python
summary_df
```

```python
summary_df["cleanliness"] = summary_df["cleanliness"] * 2
summary_df["service"] = summary_df["service"] * 2
summary_df["location"] = summary_df["location"] * 2
summary_df["value"] = summary_df["value"] * 2
summary_df["overall_rating"] = summary_df["overall_rating"] * 2
```

```python
summary_df.to_excel("output.xlsx")
```

```python
# 3. Rename columns to standardize names
summary_df.rename(columns={
    'hotel_name': 'hotel_name',
    'overall_rating': 'overall_rating',
    'total_num_of_reviews': 'total_num_of_reviews',
    'region': 'region',
    'region_shortform': 'region_shortform',
    'cleanliness': 'cleanliness',
    'service': 'service',
    'location': 'location',
    'facilities': 'facilities',
    'value': 'value'
}, inplace=True)

# Filling NaN values with NULL
summary_df.fillna('NULL', inplace=True)

# Display the resulting DataFrame
print(summary_df)
```
